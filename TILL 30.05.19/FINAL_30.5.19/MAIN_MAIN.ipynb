```python
for i in range(k):
    df['Probability{}'.format(i)]=(0.4*df1['Probability_{}'.format(i)]+0.6*df2['Probability_{}'.format(i)])

df.head(6)
```

```python
def assignment(df):
    
    counter_k = ['Probability{}'.format(i) for i in range(3)]
    #print(centroid_distance_cols)
    df['cluster_label'] = df.loc[:, counter_k].idxmax(axis=1)
    #df.loc[:, centroid_distance_cols].max(axis=1)
    df['Max_prob']=(df.loc[:, counter_k].max(axis=1))
    df['Min_prob']=(df.loc[:, counter_k].min(axis=1))
    df['cluster_label'] = df['cluster_label'].map(lambda x: int(x.lstrip('Probability')))
    #df['color'] = df['cluster_label'].map(lambda x: colmap[x]) 
    return df

assignment(df)
```

```python
def outliers(df):
    for i in range(3):
        for index,row in df.iterrows():
            z=0.25
            if(row['Min_prob']<z):
                df.drop(index,inplace=True)
                print(index)
    return df
```

```python
from sklearn.cluster import KMeans
import copy
df4=pd.read_csv('yeast_csv.csv')
df4.drop(columns='class_protein_localization',inplace=True)
for i in range(10):
    clusterer = KMeans(n_clusters=i+2, random_state=0)
    cluster_labels = clusterer.fit_predict(df4)
    silhouette_avg = silhouette_score(df4,cluster_labels)
    print(silhouette_avg,'for k =',i+2)#(-1,1) Higher the better
```
